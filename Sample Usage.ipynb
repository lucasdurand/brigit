# Diffing

```python
from yappyg import Git
git = Git('./')
```

Notebook diffs are typically quite ugly, we've installed nbdime to fix that up.

`nbdime config-git --enable --global`

```python
# normal diff (note nbdime-diff already installed/configured with
git.diff('Test.ipynb')
```

This is still not ideal, so we've wrapped a call to nbdiff-web

```python
git.nbdiff()
```
